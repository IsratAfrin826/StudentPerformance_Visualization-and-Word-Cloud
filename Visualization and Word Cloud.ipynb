```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Mount

from google.colab import drive
drive.mount('/content/drive')


# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/StudentsPerformance.csv')


# Comparative Visualization

# Calculate average scores by gender
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()

# Plot average scores by gender in a grouped bar chart
avg_scores_melted = avg_scores.melt(id_vars='gender', var_name='Subject', value_name='Average Score')


plt.figure(figsize=(10, 6))
sns.barplot(x='Subject', y='Average Score', hue='gender', data=avg_scores_melted)
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xlabel('Subject')
plt.legend(title='Gender')
plt.show()



text = ' '.join(df['parental level of education'].astype(str))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Parental Level of Education')
plt.show()
```
